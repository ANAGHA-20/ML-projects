```python
import numpy as np
import cv2 as cv
from google.colab.patches import cv2_imshow
```

```python
img = cv.imread('/content/drive/MyDrive/projects/input/Hotpot.png')
grayImage = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv2_imshow(img)
cv2_imshow(grayImage)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
low_white = (2, 5, 110)
up_white = (120, 60, 255)
# cv2_imshow(light_white)
# cv2_imshow(dark_white)
```

```python
hsv_img = cv.cvtColor(img, cv.COLOR_BGR2HSV)
mask = cv.inRange(hsv_img, lowerb = low_white, upperb = up_white)
cv2_imshow(hsv_img)
cv2_imshow(mask)
```

```python
# mask = cv.imread('mask2.png',0)
dst = cv.inpaint(img,mask,3,cv.INPAINT_TELEA)
cv2_imshow(dst)
cv.waitKey(0)
cv.destroyAllWindows()
```
